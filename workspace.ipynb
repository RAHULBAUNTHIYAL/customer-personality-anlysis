```python
Edumapdic={
    'Graduation':'Graduation','2n Cycle':'Graduation',
    'Phd':'Post Graduation','Master':'Post Graduation','Basic':'Basic'
}
Marimapdic={
    'Married' : 'Patner',
    'Together': 'Patner', 
    'Single'  : 'Single',
    'Divorce' : 'Single',
    'Widow'   : 'Single',
    'Alone'   : 'Single',
    'Absurd'  : 'Single',
    'YOLO':'Single'
}
```

```python
data['Education']=data['Education'].map(Edumapdic)
data['Marital_Status']=data['Marital_Status'].map(Marimapdic)
```

```python
data['Total spent']=data[[  'MntWines'
                            ,'MntFruits'
                            ,'MntMeatProducts'
                            ,'MntFishProducts'
                            ,'MntSweetProducts'
                            ,'MntGoldProds' ]].sum(axis='columns')
```
